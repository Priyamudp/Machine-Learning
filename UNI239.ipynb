```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

#Normalize/Scale Numeric Data
print("Normalizing/Scaling Numeric Data...")
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
print("Numeric data scaled.")
```

```python
# Step 2: Handle Date Columns
print("Handling Date Columns...")
if 'Date Recorded' in df.columns:
    df['Date Recorded'] = pd.to_datetime(df['Date Recorded'], errors='coerce')  # Convert to datetime
    df['Year Recorded'] = df['Date Recorded'].dt.year  # Extract year
    df['Month Recorded'] = df['Date Recorded'].dt.month  # Extract month
    df['Day Recorded'] = df['Date Recorded'].dt.day  # Extract day
    df.drop(columns=['Date Recorded'], inplace=True)  # Drop original date column
print("Date columns processed.")
```

```python
# Step 3: Define Features and Target
print("Defining Features and Target...")
# Define the target column (e.g., 'Sale Amount')
target_col = 'Sale Amount'  # Adjust based on your dataset
features = df.drop(columns=[target_col])  # All columns except the target
target = df[target_col]
```

```python
# Train-Test Split
print("Splitting Train and Test Data...")
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
print(f"Training set size: {X_train.shape}, Test set size: {X_test.shape}")
```

```python
# Deployment of Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Initialize and train the model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Predict on test data
y_pred_lr = linear_model.predict(X_test)

# Evaluate the model
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

# Visualization
plt.scatter(y_test, y_pred_lr, alpha=0.5)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Linear Regression: Actual vs Predicted")
plt.show()

print(f"Linear Regression - Mean Squared Error: {mse_lr}, R2 Score: {r2_lr}")
```

```python
# Deployment of Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize and train the model
decision_tree_model = DecisionTreeRegressor(random_state=42)
decision_tree_model.fit(X_train, y_train)

# Predict on test data
y_pred_dt = decision_tree_model.predict(X_test)

# Evaluate the model
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)

# Visualization
plt.scatter(y_test, y_pred_dt, alpha=0.5, color="orange")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Decision Tree: Actual vs Predicted")
plt.show()

print(f"Decision Tree - Mean Squared Error: {mse_dt}, R2 Score: {r2_dt}")
```

```python
# Deployment of Extra Trees Regressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Initialize and train the model
et_model = ExtraTreesRegressor(n_estimators=50, max_depth=10, random_state=42)  # Reduced estimators for speed
et_model.fit(X_train, y_train)

# Predict on test data
y_pred_et = et_model.predict(X_test)

# Evaluate the model
mse_et = mean_squared_error(y_test, y_pred_et)
r2_et = r2_score(y_test, y_pred_et)

# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_et, alpha=0.5, color="purple")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], '--r', linewidth=2)  # Ideal line
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Extra Trees Regressor: Actual vs Predicted")
plt.show()

print(f"Extra Trees Regressor - Mean Squared Error: {mse_et}, R2 Score: {r2_et}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np

# Define bins based on quantiles for valid intervals
bin_edges = np.quantile(y_test, [0, 0.5, 1])  # Using 0%, 50%, and 100% quantiles
y_test_binned = pd.cut(y_test, bins=bin_edges, labels=["a", "b"], include_lowest=True)
y_pred_lr_binned = pd.cut(y_pred_lr, bins=bin_edges, labels=["a", "b"], include_lowest=True)
y_pred_dt_binned = pd.cut(y_pred_dt, bins=bin_edges, labels=["a", "b"], include_lowest=True)
y_pred_et_binned = pd.cut(y_pred_et, bins=bin_edges, labels=["a", "b"], include_lowest=True)
y_pred_en_binned = pd.cut(y_pred_en, bins=bin_edges, labels=["a", "b"], include_lowest=True)

# Compute confusion matrices
confusion_matrices = {
    "Linear Regression": confusion_matrix(y_test_binned, y_pred_lr_binned, labels=["a", "b"]),
    "Decision Tree": confusion_matrix(y_test_binned, y_pred_dt_binned, labels=["a", "b"]),
    "Extra Trees": confusion_matrix(y_test_binned, y_pred_et_binned, labels=["a", "b"]),
    "ElasticNet": confusion_matrix(y_test_binned, y_pred_en_binned, labels=["a", "b"]),
}

# Display confusion matrices
for model, cm in confusion_matrices.items():
    print(f"\nConfusion Matrix for {model}:")
    print(cm)

    # Visualize the confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["a (no-recurrence-events)", "b (recurrence-events)"])
    disp.plot(cmap="Blues")
    disp.ax_.set_title(f"Confusion Matrix: {model}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Initialize a dictionary to store evaluation results
evaluation_summary = {
    "Metric": [
        "Mean Absolute Error",
        "Root Mean Squared Error",
        "Relative Absolute Error",
        "Root Relative Squared Error",
        "R2 Score",
        "Total Number of Instances"
    ]
}

# Function to calculate evaluation metrics
def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    relative_absolute_error = mae / y_true.mean()
    root_relative_squared_error = rmse / y_true.std()
    r2 = r2_score(y_true, y_pred)
    return [mae, rmse, relative_absolute_error, root_relative_squared_error, r2, len(y_true)]

# Evaluate Linear Regression
evaluation_summary["Linear Regression"] = evaluate_model(y_test, y_pred_lr)

# Evaluate Decision Tree Regressor
evaluation_summary["Decision Tree Regressor"] = evaluate_model(y_test, y_pred_dt)

# Evaluate Extra Trees Regressor
evaluation_summary["Extra Trees Regressor"] = evaluate_model(y_test, y_pred_et)

# Evaluate ElasticNet Regressor
evaluation_summary["ElasticNet Regressor"] = evaluate_model(y_test, y_pred_en)

# Convert the evaluation summary to a DataFrame
summary_df = pd.DataFrame(evaluation_summary)

# Display the summary
print(summary_df)

# Save the summary to a CSV
summary_df.to_csv("Evaluation_Summary.csv", index=False)
```
